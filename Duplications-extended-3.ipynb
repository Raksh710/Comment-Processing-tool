```python
df[df['Cosine_Similarity_Score'] > 0.8]
```

```python
def duplicate_detector(x):
    if x >= 0.8:
        return 'Identical'
    elif (x < 0.8) & (x >= 0.5):
        return 'Substantially Identical'
    else:
        return 'Unique'
```

```python
df['Similarity_Cluster'] = df['Cosine_Similarity_Score'].apply(lambda x: duplicate_detector(x))
```
