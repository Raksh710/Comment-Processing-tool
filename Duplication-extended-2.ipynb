```python
import pandas as pd, numpy as np, os
```

```python
df = pd.read_csv('Master-Similarity-2.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

```python
df[df['Cosine_Similarity_Score'] > 0.8]
```

```python
def duplicate_detector(x):
    if x >= 0.8:
        return 'Identical'
    elif (x < 0.8) & (x >= 0.5):
        return 'Substantially Identical'
    else:
        return 'Unique'
```

```python
df['Similarity_Cluster'] = df['Cosine_Similarity_Score'].apply(lambda x: duplicate_detector(x))
```

```python
df
```

```python
df['Similarity_Cluster'].value_counts()
```

```python
df[df['Similarity_Cluster']=='Unique']
```

```python
df['Similarity_Cluster'].value_counts()
```

```python
df_id = df[df['Similarity_Cluster']=='Identical'].copy()
```

```python
df_id
```

```python
id_doc_1_list = df_id['Docuemnt1'].to_list()
```

```python
id_doc_2_list = df_id['Document2'].to_list()
```

```python
id_docs = id_doc_1_list.copy()
```

```python
for i in id_doc_2_list:
    id_docs.append(i)
```

```python
df_id_2 = pd.DataFrame({'Comments':id_docs})
```

```python
df_id_2.drop_duplicates(subset='Comments', inplace=True)
```

```python
df_id_2
```

```python
df_id_2['Similarity_Cluster'] = ['Identical'] * len(df_id_2)
```

```python
df_id_2
```

```python
df_si = df[df['Similarity_Cluster']=='Substantially Identical'].copy()
```

```python
df_si
```

```python
df_si_doc1_list = df_si['Docuemnt1'].to_list()
df_si_doc2_list = df_si['Document2'].to_list()

si_docs = df_si_doc1_list.copy()

for i in df_si_doc2_list:
    si_docs.append(i)
    
df_si_2 = pd.DataFrame({'Comments':si_docs})

df_si_2.drop_duplicates(subset='Comments', inplace=True)

df_si_2['Similarity_Cluster'] = ['Substantially Identical'] * len(df_si_2)

df_si_2
```

```python
check_list_1 = df_si_2['Comments'].tolist()
```

```python
len(check_list_1)
```

```python
si_docs = []
```

```python
for i in check_list_1:
    if i not in id_docs:
        si_docs.append(i)
```

```python
len(si_docs)
```

```python
df_si_3 =pd.DataFrame({'Comments':si_docs})
```

```python
df_si_3['Similarity_Cluster'] = ['Substantially Identical'] * len(df_si_3)
```

```python
df_si_3
```

```python
df3 = pd.concat([df_id_2,df_si_3],axis=0)
```

```python
len(df3)
```

```python
df3
```

```python
df3.drop_duplicates(subset='Comments',inplace=True)
```

```python
df3
```

```python
len(df3.drop_duplicates(subset='Comments'))
```

```python
df_identical = df[df['Similarity_Cluster']=='Identical'].copy()

df_identical_doc1_list = df_identical['Docuemnt1'].to_list()
df_identical_doc2_list = df_identical['Document2'].to_list()

identical_docs = df_identical_doc1_list.copy()

for i in df_identical_doc2_list:
    identical_docs.append(i)
    
df_identical_2 = pd.DataFrame({'Comments':identical_docs})

df_identical_2.drop_duplicates(subset='Comments', inplace=True)

df_identical_2['Similarity_Cluster'] = ['Identical'] * len(df_identical_2)

df_identical_2
```

```python
df4 = pd.concat([df3,df_identical_2],axis=0)
```
